```python
%matplotlib inline
import matplotlib.pyplot as plt
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval
import numpy as np
import skimage.io as io
import os
import glob
plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
annType = ['segm','bbox','keypoints']
annType = annType[0]      #specify type here
prefix = 'person_keypoints' if annType=='keypoints' else 'instances'
print 'Running demo for *%s* results.'%(annType)
```

```python
#initialize COCO ground truth api
baseDir='/home/eitel/code/singulation_segm/self_deepmask/data/test_data'
#modelDir = 'deepmask'
#modelDir = 'sharpmask'
#modelDir = 'deepmask_nms'
#modelDir = 'sharpmask_nms'
#modelDir = 'robotpush404,remove_bg,1,revels,maxepoch_100,ft_deepmask_nms'
#modelDir = 'robotpush404_232,remove_bg,1,revels,maxepoch_100,ft_deepmask_nms'
#modelDir = 'robotpush508,remove_bg,1,revels,maxepoch_50,ft_deepmask_nms'
#modelDir = 'robotpush1329,remove_bg,1,flowfilter,15,revels,maxepoch_10,ft_deepmask,lr_0_nms'
#modelDir = 'robotpush1329,remove_bg,1,flowfilter,15,revels,maxepoch_10,ft_deepmask,lr_0,hfreq_,1.0_nms'
#modelDir = 'robotpush1329,remove_bg,1,flowfilter,0,actions,0,revels,maxepoch_10,ft_deepmask,lr_0,hfreq_1.0_nms'
#modelDir = 'robotpush1329,remove_bg,1,flowfilter,0,actions,1,revels,maxepoch_10,ft_deepmask,lr_0,hfreq_1.0_nms'

#modelDir = 'robotpush1329,remove_bg,1,flowfilter,0,actions,1,revels,maxepoch_10,ft_deepmask,lr_0,hfreq_1.0_nms'
#modelDir = 'robotpush1329,remove_bg,1,flowfilter,0,actions,0,revels,maxepoch_10,ft_deepmask,lr_0,hfreq_1.0_nms'
#modelDir = 'robotpush1329,remove_bg,0,flowfilter,0,actions,1,revels,maxepoch_10,ft_deepmask,lr_0,hfreq_1.0_nms'

#modelDir = 'robotpush1329,remove_bg,1,flowfilter,15,revels,maxepoch_10,ft_deepmask,lr_0,hfreq_,1.0_nms'
#modelDir = 'robotpush1329,remove_bg,1,flowfilter,15,actions,1,maxim,800,revels,maxepoch_10,ft_deepmask,lr_0,hfreq_1.0,trial3_nms'

# 05.2019
#modelDir = 'robotpush1329,remove_bg,1,flowfilter,15,revels,maxepoch_10,ft_deepmask,lr_0,hfreq_1.0,bootstrap0.0,trial5_nms'
modelDir = 'robotpush1329,remove_bg,1,flowfilter,15,oracle,100,revels,maxepoch_10,ft_deepmask,lr_0,hfreq_1.0,trial5_nms'
expDir = '6,8objects_aggregated_network'
if expDir == '6objects_aggregated_network':
    dataTypes=['push00','push01','push02','push03','push04','push05','push06','push07','push08']
elif expDir == '8objects_aggregated_network':
    dataTypes=['push00','push01','push02','push03','push04','push05','push06','push07','push08','push09','push10','push11']
elif expDir == '6,8objects_aggregated_network':
    dataTypes=['push00','push01','push02','push03','push04','push05','push06','push07','push08','push09','push10','push11']    

dataDir = baseDir + '/' + expDir
#Evaluate all pushes
dataType = 'push-1'
annFile = '%s/annotations/%s_%s.json'%(dataDir,prefix,dataType)
cocoGt=COCO(annFile)
resDir = '/home/eitel/code/singulation_segm/self_deepmask/pretrained/'+ modelDir + '/'+expDir
resFile = resDir + '/' + dataType +'/jsons'
print("res",resFile)
resFile =  glob.glob(resFile+"/*.json")[0]
MAP = 0
if os.path.isfile(resFile):
    print(resFile)
    cocoDt=cocoGt.loadRes(resFile)
    imgIds=sorted(cocoGt.getImgIds())
    # running evaluation
    cocoEval = COCOeval(cocoGt,cocoDt,annType)
    cocoEval.params.catIds = [1]
    cocoEval.params.imgIds  = imgIds
    cocoEval.evaluate()
    cocoEval.accumulate()
    cocoEval.summarize()
    MAP = cocoEval.stats[1]
    MAP = MAP*100
    MAP75 = cocoEval.stats[2]*100
    print("MAP50",MAP)
    print("MAP75",MAP75)
    with open(resDir+"/coco_eval.txt", "w") as f:
        print >> f, cocoEval.stats


baselines = ['deepmask_nms','sharpmask_nms',modelDir]
average_precisions = np.zeros((len(baselines),len(dataTypes)),np.float)
for b,baseline in enumerate(baselines):
    print("Baseline",baseline)
    for idx, dataType in enumerate(dataTypes):
        annFile = '%s/annotations/%s_%s.json'%(dataDir,prefix,dataType)
        cocoGt=COCO(annFile)
        # Ours
        resDir = '/home/eitel/code/singulation_segm/self_deepmask/pretrained/'+ baseline + '/'+expDir
        resFile = resDir + '/' + dataType +'/jsons'
        print("res",resFile)
        resFile =  glob.glob(resFile+"/*.json")[0]
        print("res",resFile)
        if os.path.isfile(resFile):
            print(resFile)
            cocoDt=cocoGt.loadRes(resFile)
            imgIds=sorted(cocoGt.getImgIds())
            #print("ImgIds",imgIds)
            # running evaluation
            cocoEval = COCOeval(cocoGt,cocoDt,annType)
            cocoEval.params.catIds = [1]
            cocoEval.params.imgIds  = imgIds
            cocoEval.evaluate()
            cocoEval.accumulate()
            cocoEval.summarize()
            average_precision = cocoEval.stats[1]
            average_precision = average_precision*100
            average_precisions[b,idx] = average_precision
            #print("Precision",cocoEval.evalImgs)
            #anns = cocoDt.dataset['annotations']
            #for id, ann in enumerate(anns):
            #   print("ID",id,ann)
        else:
            print("File does not exist: ",resFile)
            exit(1)

print("Average precisions",average_precisions)            

with open(resDir+"/precision_per_push.txt", "w") as f:
    print >> f, average_precisions
        
fontsz = 40
markersz = 16
linew = 2
plt.rc('font',**{'family':'cmu-serif','sans-serif':['Helvetica']})
plt.rc('text',usetex=True)
x_axis= np.linspace(0,len(dataTypes)-1, num=len(dataTypes))
print("X_axis",x_axis.shape,"Y_axis",average_precisions.shape)
plt.clf()
plt.xlabel('number of pushes',fontsize=fontsz)
plt.ylabel('AP@0.5:0.95',fontsize=fontsz)


#plt.ylim([20,80])

plt.gcf().set_size_inches(14, 9, forward=True)
plt.tick_params(axis='both', which='major', labelsize=fontsz)
#plt.hlines(average_precisions[0], 0, len(average_precisions)-1, colors='k', linestyles='--')

plt.plot(x_axis, average_precisions[0],label=baselines[0].replace('_nms',' with NMS'),marker='o', markersize=markersz, linestyle='-',linewidth=linew,color='orange')
plt.plot(x_axis, average_precisions[1],label=baselines[1].replace('_nms',' with NMS'),marker='o', markersize=markersz, linestyle='-',linewidth=linew,color='red')
plt.plot(x_axis, average_precisions[2],label='ours',marker='o', markersize=markersz, linestyle='-',linewidth=linew,color='blue')

#         ,x_axis, average_precisions[1], 'bo-', x_axis, average_precisions[2], 'go-')


#plt.plot(x_axis, average_precisions_deepmask, label='DeepMask',marker='o', markersize=markersz, linestyle='-',linewidth=linew,color='orange')


plt.legend(loc="upper left", ncol=1, shadow=False, fontsize=fontsz-6)
outfile = resDir + '/deepmask_ap.pdf'
#outfile = '/tmp/deepmask_ap.pdf'

print("Saving",outfile)
print("Mean AP",MAP)
#print("Check AP",np.mean(average_precisions))
plt.savefig(outfile)
plt.show()

    
```
